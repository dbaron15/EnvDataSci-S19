## Question 6 -  Import Into Numpy Arrays (10 pts)

Use the appropriate function to import the annual temperature and precipitation 
data into a numpy array.
***
Now that the data is downloaded, we can convert the CSVs to numpy arrays so we can use numpy functions on them with ease. This allows for fast analysis and plotting of the data.

```python
# import the data into a numpy array using the loadtxt function
annualtemp = np.loadtxt("AnnualTemperature.csv", delimiter = ",")
annualprecip = np.loadtxt("AnnualPrecipitation.csv", delimiter = ",")

# print a statement to show that the data was loaded successfully
print ("Data successfully imported into numpy arrays!")
```

## Question 7 - Print Data From Numpy Arrays (10 pts)
- Use .shape after the variable name of the numpy array (e.g. variablename.shape) to get its dimensions (i.e. number of rows and columns).  
- Print out the values of the first 10 rows and two columns of your array
- Print out the values of the last 10 rows and two columns of your array 
***
To prove that the data has been loaded successfully, we can print out information about the data and a subset of the data.

```python
# get the dimensions of the temperature and precipitation arrays
print (annualtemp.shape)
print (annualprecip.shape, "\n")

# print out the first 10 rows and two columns of each array (there are only two columns in the array so we can print all the columns)
print (annualtemp[0:10, :], "\n")
print (annualprecip[0:10, :], "\n")

# print out the last 10 rows and two columns of each array. use the last index 
print (annualtemp[139:149, :], "\n")
print (annualprecip[139:149, :])
```

## Question 8 - Run Calculations on Numpy Arrays (10 pts)

- Convert the variable values from Fahrenheit to Celsius using the following equation:
Celsius = (Fahrenheit - 32) / 1.8
Precipitation (mm) = precipitaton (in)*25.4
- Print out the values of the first 10 rows and two columns of your array 
- Print out the values of the last 10 rows and two columns of your array 
***
Now that the data is successfully imported into arrays, we can run calculations on the arrays, like converting the temperatures and precipitations into metric measurements.

```python
# convert the temperature array to Celsius from Fahrenheit
annualtemp[:,1] = (annualtemp[:,1] - 32) / 1.8

# print out the first and last 10 rows and two columns of the temperature array
print (annualtemp[0:10, :], "\n")
print (annualtemp[139:149, :], "\n")
```

```python
# convert the precipitation array from inches to millimeters
annualprecip[:,1] = (annualprecip[:,1] * 25.4)

# print out the first and last 10 rows and two columns of the precipitation array
print (annualprecip[0:10, :], "\n")
print (annualprecip[139:149, :], "\n")
```

## Question 9 - Calculate Summary Statistics of Numpy Array (10 pts)
- Calculate the minmum, maximum, and mean and standard deviation of the annual temperature and precipitation data 
***
We can also calculate basic statistics on the arrays for better summarization of the data.

```python
# calculate the min, max, mean, and standard deviation of the temperature data
temp = annualtemp[:,1] # convert to single array to allow for easier processing
mean_temp = np.mean(temp)
min_temp = np.min(temp)
max_temp = np.max(temp)
std_temp = np.std(temp)

# print out the values for temperature
print ("The average annual temperature is", mean_temp, "degrees Celsius.\n")
print ("The minimum temperature is", min_temp, "degrees Celsius.\n")
print ("The maximum temperature is", max_temp, "degrees Celsius.\n")
print ("The standard deviation of the temperature is", std_temp, "degrees Celsius.\n")
```

```python
# calculate the min, max, mean, and standard deviation of the precipitation data
precip = annualprecip[:,1] # convert to single array to allow for easier processing
mean_precip = np.mean(precip)
min_precip = np.min(precip)
max_precip = np.max(precip)
std_precip = np.std(precip)

# print out the values for temperature
print ("The average annual precipitation is", mean_precip, "mm.\n")
print ("The minimum precipitation is", min_precip, "mm.\n")
print ("The maximum precipitation is", max_precip, "mm.\n")
print ("The standard deviation of the precipitation is", std_precip, "mm.\n")
```

Now we can plot the data to better communicate it to other audiences visually.

```python
# Let's plot the numpy arrays on the same subplot!!

# set the plot size for these plots
plt.rcParams["figure.figsize"] = (12, 12)

# create the plot space with two subplots on it
fig, (ax1, ax2) = plt.subplots(2) # default `sharex = False`, it could be true because they are the same, I don't like how it looks

# add the x and y axes to the graph (scatter and line)
ax1.plot(annualtemp[:,0], annualtemp[:,1], "-o", color = "red")
ax2.plot(annualprecip[:,0], annualprecip[:,1], "-o", color = "blue")

# set the plot titles
ax1.set(title = "Time Series of Annual Temperature in Central Park, NY from 1868 to 2017 in Celcius")
ax2.set(title = "Time Series of Annual Precipitation in Central Park, NY from 1868 to 2017 in Millimeters")

# set the axes titles
ax1.set(xlabel = "Year", ylabel = "Temperature (degrees Celsius)")
ax2.set(xlabel = "Year", ylabel = "Precipitation (mm)");
```
