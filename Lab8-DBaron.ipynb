In the way I've written my function, making the plots in a loop and a function are relatively similar. Setting the figure, titles, and legend outside the function allows the function to write to the figure and overplot the two datasets. This is true for the loop and the function, because it goes through each CSV file separately and in its entirety. To overplot, it makes sense to set a figure to write to. With the loop, you only need to set the figure and the subplots, and the loop will allow you to write the axis labels, the title, and the legend for each input. In the function, it makes more sense to set the titles and legend outside the function, as it will be the same for both runs of the function.
