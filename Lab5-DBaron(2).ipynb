## Question 7 - Run Calculations on pandas dataframes (10 pts)

- Convert the temperature from Fahrenheit to Celsius using the following equation:
Celsius = (Fahrenheit - 32) / 1.8
- Convert the precipitation from inch to mm using mm = inch*25.4
- Print out the values of the first 10 rows and two columns of your array 
- Print out the values of the last 10 rows and two columns of your array 

```python
# convert the temperature dataframe from Fahrenheit to Celsius using indices
temp.iloc[:, 1:14] = (temp.iloc[:, 1:14] - 32) / 1.8
```

```python
# print the first 10 rows and 2 columns of the new temperature array
print(temp.iloc[0:10, 0:2])
```

```python
# print the last 10 rows and 2 columns of the new temperature dataframe
print(temp.iloc[-10:, 0:2])
```

```python
# convert the precipitation dataframe from inches to millimeters
precip.iloc[:, 1:14] = (precip.iloc[:, 1:14] * 25.4)
```

```python
# print the first 10 rows and 2 columns of the new precipitation dataframe
print(precip.iloc[0:10, 0:2])
```

```python
# print the last 10 rows and 2 columns of the new precipitation array
print(precip.iloc[-10:, 0:2])
```

## Question 9 - Plot the boxplot of precip and tempt.  (15 pts)
- plot the boxplot of monthly and annual precip and tempt as two subplots in one plot

```python
# set the boxplots for the temperature and precipitation dataframes on the same subplot

# set plot size for all plots
plt.rcParams["figure.figsize"] = (16, 16)

# set two subplots for the figure
fig, (ax1, ax2) = plt.subplots(2)

# load the temperature boxplot
temp.iloc[:, 1:13].boxplot(ax=ax1, whis="range")

# set plot title
ax1.set(title="1869 - 2017 Statistics of Monthly Temperature in Central Park, NY")

# load the precipitation boxplot
precip.iloc[:, 1:13].boxplot(ax=ax2, whis="range")

# set second plot title
ax2.set(title="1869 - 2017 Statistics of Monthly Precipitation in Central Park, NY");
```

## Bonus question: can you plot the annual precipitation variation before and after 1960 (10pts)

```python
# set year as index column
precip = precip.set_index("YEAR")

# get the data before and after 1960 using loc index
b_1960 = precip.loc[:"1959"]
a_1960 = precip.loc["1960":]
```

```python
# plot the boxplots for before and after 1960

# set plot size for all plots
plt.rcParams["figure.figsize"] = (12, 8)

# set two subplots for the figure
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1)

# load the before 1960 boxplot (ANNUAL column only)
b_1960.iloc[:, 12:13].boxplot(ax=ax1)

# set plot title
ax1.set(title="Statistics of Annual Precipitation Before 1960 \n in Central Park, NY")

# load the after 1960 boxplot (ANNUAL column only)
a_1960.iloc[:, 12:13].boxplot(ax=ax2)

# set second plot title
ax2.set(title="Statistics of Monthly Precipitation After 1960 \n in Central Park, NY");
```
